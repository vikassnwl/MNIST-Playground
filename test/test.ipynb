```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd
```

```python
model_cnn = tf.keras.models.load_model("../model/mnist_model_cnn")
```

```python
testData = pd.read_csv("testData.csv")
```

```python
testData
```

```python
fig, axes = plt.subplots(2, 5, figsize=(10, 4))
for i, row in testData.iterrows():
    axis = axes[i//5][i%5]
    image = row.to_numpy().reshape(28, 28)/255.
    probas = model_cnn.predict(np.array([image]), verbose=0)[0]
    pred = probas.argmax()
    proba = probas[pred]
    axis.imshow(image, cmap="gray")
    axis.set_title(f"{pred}, {proba:.2f}")
    axis.axis("off")
```
